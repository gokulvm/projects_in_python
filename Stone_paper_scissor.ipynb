# Stone Paper Scissor Game

```python
from numpy import random
print("welcome to stone paper scissor game")
print("This game is between player vs computer")
print("This game give you to 3 chance,who won maximum times,he is the winner")

game = ["stone","paper","scissor"]


count = 1
player = 0
com = 0

while count <= 3:
    computer = game[random.randint(0, 3)]   #computer choise
    print("""1 - stone
2 - paper 
3 - scissor""")

    man = int(input("Enter your decision : ")) #getting input from the player
    man = man-1

    if man >= 0 and man < 3:
        human = game[man]
        if human == computer:
            print("Its a tie")
            print("you need to play extra one time")
            continue   #if game was tie try again

        elif human == "stone":
            count +=1
            if computer == "paper":
                com += 1

            elif computer == "scissor":
                player += 1

        elif human == "paper":
            count += 1
            if computer == "stone":
                player += 1

            elif computer == "scissor":
                com +=1

        elif human == "scissor":
            count += 1
            if computer == "stone":
                com +=1

            elif computer == "paper":
                player +=1

    else:  #if player enter wrong input tell them to try again
        print("Enter a valid number")
        continue


#printing the winners
if com < player:
    print("You won the match")

else:
    print("You lose the match "
          "Better luck next time")

```
