# Tik Tac Toe Game

```python
board = ["-","-","-","-","-","-","-","-","-"]   #Board
winner = None  #This variable used to find who is the winner
game_playing = True


def game():
    while game_playing:  #This loop run until game_playing become false
        print_board()
        player_input()
        check_game_tie()
        game_over()
        flip_player()

    if winner == "X" or winner == "O":  #after game over print winner
        print_board()
        print(f"{winner} won")

    elif winner==None:  #if game was tie
        print_board()
        print("Game was Tie")



def print_board(): #display the board
    print(f"{board[0]} {board[1]} {board[2]}")
    print(f"{board[3]} {board[4]} {board[5]}")
    print(f"{board[6]} {board[7]} {board[8]}")



def player_input():  #getting input from the player
    check = True
    global inpt
    while check:
        print("enter the value between 1-9")
        inpt = input(f"enter where you want to place your {player} : ")

        while inpt in ["1","2","3","4","5","6","7","8","9"]: #check the input value is 1 to 9
            inpt = int(inpt) - 1
            inpt = int(inpt)
            if board[inpt] == "-" and board[inpt] != "X" and "O":   #check player entered place already filled or not
                check = False
            else:
                print("this place is already filled select other places")


    board[inpt] = player



player = "X"
def flip_player():   #This function change the player X to O
    global player
    if player == "X":
         player = "O"
    elif player == "O":
        player = "X"


def game_over():
    global winner
    row = row_checks()
    col = column_check()
    diag = diagonal_check()

    if row:
        winner = row
    elif col:
        winner = col
    elif diag:
        winner = diag
    else:
        winner = None

    return


def row_checks():  #checking winner in row wise
    global game_playing
    r1 = board[0] == board[1] == board[2] != "-"
    r2 = board[3] == board[4] == board[5] != "-"
    r3 = board[6] == board[7] == board[8] != "-"
    if r1 or r2 or r3:  #if winner is found stop game_playing execustion
        game_playing = False

    if r1:  #This will be return winner is X or O
        return board[0]
    elif r2:
        return board[3]
    elif r3:
        return board[8]


def column_check():    #check winner column wise
    global game_playing

    c1 = board[0] == board[3] == board[6] != "-"
    c2 = board[1] == board[4] == board[7] != "-"
    c3 = board[2] == board[5] == board[8] != "-"
    if c1 or c2 or c3 :
        game_playing = False
    if c1:
        return board[0]
    elif c2:
        return board[1]
    elif c3:
        return board[2]


def diagonal_check():    #check winner in diagonal wise
    global game_playing

    d1 = board[0] == board[4] == board[8] != "-"
    d2 = board[2] == board[4] == board[6] != "-"
    if d1 or d2:
        game_playing = False
    if d1:
        return board[0]
    elif d2:
        return board[2]


def check_game_tie():
    global game_playing
    global winner

    if "-" not in board:  #if "-" this value is not in the board the game was tie.
        game_playing = False
        winner = None
    return winner


game()
```
